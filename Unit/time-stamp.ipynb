```python
import cv2

cap = cv2.VideoCapture('video.mp4')  # Ganti dengan lokasi video Anda

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    timestamp_milliseconds = cap.get(cv2.CAP_PROP_POS_MSEC)
    timestamp_seconds = timestamp_milliseconds / 1000.0
    
    print("Timestamp (milliseconds):", timestamp_milliseconds)
    print("Timestamp (seconds):", timestamp_seconds)
    
    cv2.imshow('Frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

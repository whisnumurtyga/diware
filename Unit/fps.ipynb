```python
import cv2
from datetime import datetime
import time

# Open the video source (0 for default camera)
cap = cv2.VideoCapture(0)

# Set the desired FPS (e.g., 30 FPS)
desired_fps = 16
cap.set(cv2.CAP_PROP_FPS, desired_fps)

i = 0
prev_time = time.time()
frame_count = 0
while True:
    start = datetime.now()
    ret, frame = cap.read()
    if not ret:
        break
    
    # Get the actual FPS
    # Process the frame here
    
    frame_count += 1
    current_time = time.time()
    time_elapsed = current_time - prev_time
    
    
    
    
    # Process the frame here
    if time_elapsed >= 1:
        actual_fps = frame_count / time_elapsed
        
        # Draw actual FPS on the frame
        fps_text = f"Actual FPS: {actual_fps:.2f}"
        cv2.putText(frame, fps_text, (10, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        prev_time = current_time
        frame_count = 0
    cv2.imshow('Frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    i+=1
    end = datetime.now()
    print(f'frame {i}\t{(end-start).total_seconds()} detik')
    

cap.release()
cv2.destroyAllWindows()
```

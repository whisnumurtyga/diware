```python
'''
    Import Library yang Dibutuhkan
'''
import mediapipe as mp
import numpy as np
import cv2
```

```python
'''
    - Baca Input dari Kamera Puh
    - Membuat objek face_mesh
    - Konfigurasi Objek face_mesh (facmesh)
    - Impor modul drawing_utils untuk styling landmark
    
'''
cap = cv2.VideoCapture(0)
facmesh = mp.solutions.face_mesh
face = facmesh.FaceMesh(
        static_image_mode = True, 
        min_tracking_confidence = 0.6,
        min_detection_confidence = 0.6,
    )
draw = mp.solutions.drawing_utils
''' 
    Proses Per Framenya Puh
'''
while True:
    '''
        - Baca per frame
        - Simpan h dan w untuk denormalize koordinat
        - Image yang ditangkap format BGR (dari OpenCV) ubah ke RGB
        - Menerapkan model face_mesh ke frame yang udah diubah ke RGB (rgb) disimpan di variabel (op)
        - Cek Apakah ada face landmarks yang terdeteksi?
            - Iterasi setiap titik landmarksnya (i)
            - Gambar Landmarknya di var(frm), dengan koordinat x-y-z (i), sambungkan titiknya dengan FACEMESH_CONTOURS,
                - Konfigurasi radius circle dan warna dari titik landmarks
            - Print Koordinat sebelum di denormalize dan normalize 
        - Tampilkan window dari var (frm)
    '''
    _, frm= cap.read()
    h, w, _ = frm.shape
    rgb = cv2.cvtColor(frm, cv2.COLOR_BGR2RGB)
    
    op = face.process(rgb)
    if op.multi_face_landmarks:
        for i in op.multi_face_landmarks:
            ''' 
                Alternatif apabila facmesh.FACE_CONNECTIONS Error => FACEMESH_COUNTOURS / FACE_CONTOURS
            '''
            draw.draw_landmarks(frm, i, facmesh.FACEMESH_CONTOURS,
                                landmark_drawing_spec=draw.DrawingSpec(
                                        circle_radius=1,
                                        color=(28, 214, 23)
                                    
                                    )
                                )
            print('-== Denormalize Coordinates ==-\t\t\t-== Normalize Coordinate ==-')
            print(f'x: {i.landmark[0].x * w}\t\t\t{i.landmark[0].x}')
            print(f'x: {i.landmark[0].y * h}\t\t\t{i.landmark[0].y}')
            print(f'x: {i.landmark[0].z * h}\t\t\t{i.landmark[0].z}')
            
    cv2.imshow('window', frm)
    
    '''
        Ini Untuk konfigurasi exit puh
        ord('q') => tombol huruf 'q' pada keyboard
        27 => tombol esc
    '''
    if cv2.waitKey(1) == ord('q') or cv2.waitKey(1) == 27:
        cap.release()
        cv2.destroyAllWindows()
        break
```
